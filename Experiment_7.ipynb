```python
# while loop
# Used when number of iterations is not known
# Print i as long as i is less than 6 :
i = 1
while i<=5 :
  print(i)
  i += 1
```

```python
# Print numbers from 1 to n
n = int(input("Enter n:"))
i = 1
while i<=n:
  print(i)
  i+= 1
```

```python
# Factorial of a number
n = int(input("Enter a number: "))
fact = 1
while n>0:
  fact *=n
  n-= 15

print("Factorial =", fact)
```

```python
# Fibonacci
n = int(input("Enter number of terms : "))
a = 0
b = 1
i = 1
while i<=n :
  print(a, end = " ")
  c = a + b
  a = b
  b = c
  i += 1
```

```python
# Limit
limit = int(input("Enter limit : "))
a = 0
b = 1
while a <= limit :
  print(a, end = " ")
  a,b = b, a + b
```

```python
# Reverse of a number
n = int(input("Enter a number: "))
rev = 0
while n > 0:
  digit = n % 10
  rev = rev *10 + digit
  n //= 10
print("Reversed number = ", rev)
```

```python
# Check Palindrome for a number
num = int(input("Enter a number: "))
x = num
rev = 0
while num > 0:
  rev = rev * 10 + num % 10
  num //= 10
if x == rev :
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Check Palindrome for a string
s = str(input("Enter a string: "))
i,j = 0,len(s) - 1
isp = True
while i < j:
  if s[i] != s[j]:
     isp = False
     break
  i += 1
  j -= 1
if isp:
  print("Yes")
else:
  print("No")
```

```python
# Check for a string shorter method
s = input("Enter the string: ")
rev = s[::-1]
if s == rev:
  print("It is a Palindrome ")
else :
  print("Not a Palindrome ")
```

```python
# Count digits in a number
num = int(input("Enter a number: "))
count = 0
while num > 0:
  count +=1
  num = num // 10

print("Number of digits = " , count)
```

```python
# Search for an element in a list
num = [10,20,30,40,50]
key = int(input("Enter the element you want to search :" ))

i = 0
while i < len(num):
  if num[i] == key:
   print("Element found at index", i)
   break
  i += 1
else:
  print("Element not found")
```

```python
# Print only odd numbers between 1 to 10
i = 0
while i < 10:
  i += 1
  if i % 2 == 0:
    continue
  print(i)
```
